```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 8))
features_to_analyze = ['streams', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%']
for i, feature in enumerate(features_to_analyze, 1):
    plt.subplot(3, 2, i)
    sns.histplot(data[feature], kde=True, bins=30, color='b')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
data.dtypes
```

```python
print(data['streams'].head())

# Step 2: Remove non-numeric characters (e.g., commas) from 'streams'
data['streams'] = data['streams'].str.replace(',', '')

# Step 3: Convert the 'streams' column to numeric (handle invalid values)
data['streams'] = pd.to_numeric(data['streams'], errors='coerce')

# Step 4: Convert the 'streams' column to int64 (after handling NaN values)
data['streams'] = data['streams'].astype('int64', errors='ignore')

# Verify the data type has been changed
print(data['streams'].dtypes)
```

```python
data['streams'].isnull().sum()
```

```python
data['streams'].fillna(data['streams'].median())
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = data[['streams', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='released_year', y='streams', data=data, estimator='sum', errorbar=None, marker='o', color='green')
plt.title('Total Streams per Year')
plt.xlabel('Year')
plt.ylabel('Total Streams')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_artists = data['artist(s)_name'].value_counts().head(10)
sns.barplot(x=top_artists.index, y=top_artists.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Top 10 Artists by Song Count')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='acousticness_%', y='energy_%', data=data, hue='streams', size='streams', sizes=(40, 400), palette="cool")
plt.title('Energy vs Acousticness (Size by Streams)')
plt.xlabel('Acousticness (%)')
plt.ylabel('Energy (%)')
plt.legend(title='Streams', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_songs = data[['track_name', 'streams']].sort_values(by='streams', ascending=False).head(10)
# Assign 'track_name' to hue and set legend=False
sns.barplot(x='streams', y='track_name', data=top_songs, hue='track_name', palette='magma', legend=False)
plt.title('Top 10 Songs by Streams')
plt.xlabel('Streams')
plt.ylabel('Song')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_features = ['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']
df_scaled = pd.DataFrame(scaler.fit_transform(data[numeric_features]), columns=numeric_features)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
data['pca_one'] = pca_result[:, 0]
data['pca_two'] = pca_result[:, 1]
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='pca_one', y='pca_two', data=data, hue='streams', palette='cool', legend=False)
plt.title('PCA of Song Attributes')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
data['cluster'] = kmeans.fit_predict(df_scaled)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='pca_one', y='pca_two', data=data, hue='cluster', palette='tab10', legend='full')
plt.title('K-Means Clustering of Songs')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
